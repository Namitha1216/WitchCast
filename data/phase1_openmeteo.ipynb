```python
#CELL 1
import requests
import pandas as pd

# Example: Dubai
lat, lon = 25.276987, 55.296249
url = f"https://api.open-meteo.com/v1/forecast?latitude={lat}&longitude={lon}&daily=temperature_2m_max,temperature_2m_min,precipitation_sum,windspeed_10m_max&timezone=auto"

resp = requests.get(url)
data = resp.json()

df = pd.DataFrame(data["daily"])
print(df.head())

# Simple classification
def classify(row):
    if row["temperature_2m_max"] > 38:
        return "Very Hot"
    elif row["temperature_2m_min"] < 5:
        return "Very Cold"
    elif row["windspeed_10m_max"] > 50:
        return "Very Windy"
    elif row["precipitation_sum"] > 10:
        return "Very Wet"
    else:
        return "Comfortable"

df["category"] = df.apply(classify, axis=1)
df.to_csv("phase1_forecast.csv", index=False)
df
```
